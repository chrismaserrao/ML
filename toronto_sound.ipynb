```python
## function for normalizing the dataset
```

```python
def normalize_audio(filename):
    filename = filename / np.max(np.abs(filename))
    return filename
```

```python
filename = normalize_audio(filename)
plt.figure(figsize=(15,4))
plt.plot(np.linspace(0, len(filename) / sample_rate, num=len(filename)), filename)
plt.grid(True)
```

```python
sample_rate ## how many times per second a sound has sampled
## standard is 44.1 khz and (44,100)
## scipy lib can also be used, 
## two channels serial- diff in right and left hand side, and mono - headset
```
